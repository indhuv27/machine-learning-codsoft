MODEL TRAINING AND EVALUATION

```python
 # basedon the dataset , these columns contain non-numeric data or a identifiers not useful for the model
columns_to_drop = ['Unnamed: 0', 'cc_num', 'merchant', 'category', 'first', 'last', 'street', 'city', 'state', 'zip', 'job', 'dob', 'trans_num', 'trans_date_trans_time'] # Add any other non-numerical columns as needed

# Drop the specified columns from the new_dataset to create the feature set X
# Also drop the 'is_fraud' column to create the features
X = new_dataset.drop(columns=['is_fraud'] + columns_to_drop, axis=1)

# The target variable Y remains 'is_fraud'
Y = new_dataset['is_fraud']

# Identify non-numeric columns remaining in X
non_numeric_columns_remaining = X.select_dtypes(exclude=['number']).columns

# Apply one-hot encoding to the remaining non-numeric columns (like 'gender')
X = pd.get_dummies(X, columns=non_numeric_columns_remaining, drop_first=True) # drop_first=True avoids multicollinearity

# %%
# print(X)
# %%
# print(Y)
# %%
# %%
# split the data into training and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
# %%
print(X.shape,X_train.shape,X_test.shape)
# %% [markdown]
# MODEL TRAINING
# %%

# create an instance of the LogisticRegression model
model = LogisticRegression()
# training the model
model.fit(X_train,Y_train)
# %% [markdown]
# MODEL EVALUATION
#
#
# ACCURACY SCORE
# %%
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)

# print the accuracy score on the training data
print('Accuracy on Training Data:', training_data_accuracy)

# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)

# print the accuracy score on the test data
print('Accuracy on Test Data:', test_data_accuracy)
```
